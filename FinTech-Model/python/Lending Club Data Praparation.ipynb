### Format conversion of feature names

```python
# current data全部改成小写
```

```python
df_current.columns.values
```

```python
current_feature = list(df_current.columns.values)
```

```python
for i in range(len(current_feature)):
    current_feature[i] = str.lower(str(current_feature[i]))
```

```python
current_feature[:5]
```

```python
#historical data全部去掉下划线
```

```python
df_2014.columns.values
```

```python
hist_feature = list(df_2014.columns.values)
```

```python
for i in range(len(hist_feature)):
    hist_feature[i] = str(hist_feature[i].replace('_', ''))
```

```python
hist_feature[:5]
```

```python
#利用集合找到共同的features
```

```python
common_feature = set(hist_feature)&(set(current_feature))
print(len(common_feature))
print(common_feature)
```

```python
hist_unique = set(hist_feature)-set(current_feature)
print(len(hist_unique))
print(hist_unique)
```

```python
current_unique = set(current_feature)-set(hist_feature)
print(len(current_unique))
print(current_unique)
```

```python
# add common features that have different names for current and historical data. use the form in the historical data
```

```python
# historical vs. current data
# loanamnt vs. loanamount
# zipcode vs. addrzip
# fundedamnt vs. fundedamount
# numacctsever120pd vs. numacctsever120ppd
# verificationstatus vs. isincv
# verificationstatusjoint isincvjoint

# lastcreditpulld vs. creditpulld，贷款过程完成后的记录，可能不同于current data中的（贷款申请时的）记录
# issued vs. acceptd, current data均为待审批的贷款记录，没有issued
```

```python
common_feature = common_feature.union({'zipcode',
                                      'fundedamnt',
                                      'verificationstatus',
                                       'verificationstatusjoint',
                                      'loanamnt',
                                      'numacctsever120pd'})
len(common_feature)
```
